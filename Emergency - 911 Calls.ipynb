# 911 Calls Capstone Project

This project is an in-depth analysis of 911 call data. The goal is to uncover insights into the nature of emergencies reported, their geographical distribution, and temporal trends. This analysis could be beneficial for emergency response planning and public safety strategies.


## Methodology
The analysis is conducted using Python, with libraries such as Pandas for data manipulation, Matplotlib and Seaborn for visualization. The dataset comprises records of 911 emergency calls, including details like time of call, reason for the call, and location.

** Importing Libraries **

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Reading CSV

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head(5)
```

```python
df['twp'].value_counts().head(5)
```

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df,palette='deep')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
sns.countplot(x='Day of Week', data=df, hue='Reason', palette='deep')
```

```python
sns.countplot(x='Month', data=df, hue='Reason', palette='deep')
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda t: t.date())
```

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
df[df['Reason'] == 'Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason'] == 'Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason'] == 'EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
sns.heatmap(dayHour, cmap = 'coolwarm')
```

```python
sns.clustermap(dayHour, cmap = 'coolwarm')
```

```python
dayHour = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayHour.head()
```

```python
sns.heatmap(dayHour, cmap = 'coolwarm')
```

```python
sns.clustermap(dayHour, cmap = 'coolwarm')
```
